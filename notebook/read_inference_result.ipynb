# Read inference results

This notebook shows how to:
- load a test-set inference result file such as `results/test_set_inference/unitslib_formula_oos/truth_pred_hiegnn.npy`
- inspect its dictionary structure
- export reference and generated molecules to `.xyz` files for visualization


```python
import warnings; warnings.filterwarnings("ignore")
from pathlib import Path
import numpy as np
from units.generate import load_inference_results, export_inference_sample_xyz, export_inference_batch_xyz

result_path = "../results/test_set_inference/unitslib_formula_oos/truth_pred_hiegnn.npy"
batch_idx = 10
sample_idx = 1
repeat_idx = 1
results = load_inference_results(result_path)

print(f"Loaded: {result_path}")
print(f"Number of batches: {len(results)}")
print(f"Batch indices: {list(results.keys())}")
```

```python
batch_result = results[batch_idx]

print("batch keys:", list(batch_result.keys()))
print("batch_idx:", batch_result["batch_idx"])
print("len(truth_xyz):", len(batch_result["truth_xyz"]))
print("pred_xyz shape:", np.asarray(batch_result["pred_xyz"]).shape)
print("node_mask shape:", np.asarray(batch_result["node_mask"]).shape)
print("mol_atoms shape:", np.asarray(batch_result["mol_atoms"]).shape)
```

The stored structure is:

- `results[batch_idx]` is a dictionary with keys `['batch_idx', 'truth_xyz', 'pred_xyz', 'node_mask', 'mol_atoms']`
- `pred_xyz.shape == (repeat_time, batch_size, max_atom_num, 3)`
- `truth_xyz[i]` stores the reference coordinates for molecule `i` in the batch
- `node_mask[i]` selects valid atoms from padded arrays


```python
# Example 1: export one molecule in one batch
output_dir = Path("./exported_xyz")

exported_paths = export_inference_sample_xyz(
    results,
    batch_idx=batch_idx,
    sample_idx=sample_idx,
    output_dir=output_dir,
    repeat_idx=repeat_idx,
    export_all_repeats=False,
)

for path in exported_paths:
    print(path)
```

```python
# Example 2: export all generated repeats for one molecule
exported_paths = export_inference_sample_xyz(
    results,
    batch_idx=batch_idx,
    sample_idx=sample_idx,
    output_dir=output_dir,
    export_all_repeats=True,
)

print(f"Exported {len(exported_paths)} xyz files.")
```
